```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
from matplotlib import rcParams
rcParams['figure.figsize'] = 13,9
```

###  Retrieve Data

```python
data=pd.read_excel('INX_Future_Inc_Employee_Performance_CDS_Project2_Data_V1.8 - Copy.xls')
data.head()
```

```python
data.plot()
```

# Department-Wise Performance Rating Visualization

```python
dept_wise = data[['EmpDepartment','PerformanceRating']]
dept_wise.head()
```

```python
dept = pd.crosstab(dept_wise.EmpDepartment,dept_wise.PerformanceRating)
dept
```

```python
x = np.arange(len(dept['EmpDepartment']))
x
```

```python
plt.xticks(x,dept['EmpDepartment'],color='r',rotation =60, fontsize=10)
plt.bar(x-0.2, dept[2],width=0.3,label='Good')
plt.bar(x, dept[3],width=0.3,label='Excellent')
plt.bar(x+0.2, dept[4],width=0.3,label='Outstanding')
plt.legend(loc='upper right',shadow=True)
plt.xlabel('Department',fontsize=14)
plt.ylabel('Number of people',fontsize=14)
plt.title('Department Wise Performance',fontsize=18)
rcParams['figure.figsize'] = 13,9
plt.show()
```

# Top 3 Factors Effecting  Employee Performance Visualization

### 1.Employee Environment Satisfaction
### 2.Employee Last Salary Hike Percentage
### 3.Employee Work Life Balance

## 1.Employee Environment Satisfaction visualization

```python
EmpES= pd.crosstab(data.EmpEnvironmentSatisfaction,dept_wise.PerformanceRating)
EmpES
```

```python
EmpES.reset_index(inplace=True)
EmpES
```

```python
x = np.arange(len(EmpES['EmpEnvironmentSatisfaction']))
x
```

```python
plt.xticks(x,['Low','Medium','High','Very-High'])
plt.bar(x-0.14, EmpES[2],width=0.3,label='Good')
plt.bar(x+0.14, EmpES[4],width=0.3,label='Outstanding')
plt.legend(loc='upper right',shadow=True)
plt.xlabel('Employee Environment Satisfaction',fontsize=12)
plt.ylabel('Number of people',fontsize=12)
plt.title('Employee Environment Satisfaction Wise Performance',fontsize=16)
plt.show()
```

## 2.Employee Last Salary Hike Percent Visualization

```python
EmpLS = pd.crosstab(data.EmpLastSalaryHikePercent,dept_wise.PerformanceRating)
EmpLS
```

```python
EmpLS.reset_index(inplace=True)
EmpLS
```

```python
x = np.arange(len(EmpLS['EmpLastSalaryHikePercent']))
x
```

```python
plt.xticks(x,EmpLS['EmpLastSalaryHikePercent'])
plt.bar(x-0.1,EmpLS[2],width=0.3,label='Good')
plt.bar(x+0.1, EmpLS[4],width=0.3,label='Outstanding')
plt.legend(loc='upper right',shadow=True)
plt.xlabel('Employee Last Salary Hike Percentage',fontsize=12)
plt.ylabel('Number of people',fontsize=12)
plt.title('Last Salary Hike percent Wise Performance',fontsize=16)
plt.show()
```

## 3.Employee Work Life Balance Visualization

```python
EmpWL= pd.crosstab(data.EmpWorkLifeBalance,dept_wise.PerformanceRating)
EmpWL
```

```python
EmpWL.reset_index(inplace=True)
EmpWL
```

```python
x = np.arange(len(EmpWL['EmpWorkLifeBalance']))
```

```python
plt.xticks(x,['Bad','Good','Better','Best'])
plt.bar(x-0.15, EmpWL[2],width=0.3,label='Good')
plt.bar(x+0.15, EmpWL[4],width=0.3,label='Outstanding')
plt.legend(loc='best',shadow=True)
plt.xlabel('Employee Work Life Balance',fontsize=12)
plt.ylabel('Number of people',fontsize=12)
plt.title('Work Life Balance Wise Performance',fontsize=16)
plt.show()
```
